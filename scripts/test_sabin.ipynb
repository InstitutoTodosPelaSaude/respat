### **Creating renaming Dicts**

```python
## import reduce function
from functools import reduce

rename_cols_dict = (
    df_rename_cols
    .query('lab_id == "SABIN"')
    .assign( mapping=lambda df: df.apply(lambda row: {row['column_name']: row['new_name']}, axis=1) )
    .groupby('lab_id')
    .agg({'mapping': lambda x: reduce(lambda a,b: {**a,**b}, x)})
    .to_dict()
    ['mapping']
)

rename_cols_dict
```

### **Fixing Columns**

```python
df_sabin_covid = pd.read_excel(
    './../data/SABIN/20230711_SABIN_Covid_2023ateSE27.xlsx'
)
```

**Fixing Column Types**

```python
df_sabin_covid["OS"] = df_sabin_covid["OS"].astype('str')
df_sabin_covid["Código Posto"] = df_sabin_covid["Código Posto"].astype('int16')
df_sabin_covid["Estado"] = df_sabin_covid["Estado"].astype('str')
df_sabin_covid["Municipio"] = df_sabin_covid["Municipio"].astype('str')
df_sabin_covid["DataAtendimento"] = pd.to_datetime(df_sabin_covid["DataAtendimento"])
df_sabin_covid["DataNascimento"] = pd.to_datetime(df_sabin_covid["DataNascimento"])
df_sabin_covid["Sexo"] = df_sabin_covid["Sexo"].astype('str')
df_sabin_covid["Descricao"] = df_sabin_covid["Descricao"].astype('str')
df_sabin_covid["Parametro"] = df_sabin_covid["Parametro"].astype('str')
df_sabin_covid["Resultado"] = df_sabin_covid["Resultado"].astype('str')
df_sabin_covid["DataAssinatura"] = pd.to_datetime(df_sabin_covid["DataAssinatura"])
```

**Adding Sample ID and Test Kit information**

```python
df_sabin_covid.insert(1, 'sample_id', '')
df_sabin_covid.insert(1, 'test_kit', '')
df_sabin_covid["test_kit"] = df_sabin_covid["Parametro"].apply(lambda x: "covid_antigen" if x == "COVIDECO" else "covid_pcr") # separate by rows in column test_kit
df_sabin_covid.fillna('', inplace=True)
```

**Handling ID Columns**

```python
id_columns = [
    'OS',
    'Estado',
    'Municipio',
    'DataAtendimento',
    'Sexo',
    'Descricao',
    'Resultado',
]

for column in id_columns:
    if column not in df_sabin_covid.columns.tolist():
        df_sabin_covid[column] = ''
        print('\t\t\t - No \'%s\' column found. Please check for inconsistencies. Meanwhile, an empty \'%s\' column was added.' % (column, column))
```

```python
if 'DataNascimento' not in df_sabin_covid.columns.tolist():
    df_sabin_covid['birthdate'] = ''
df_sabin_covid['Ct_FluA'] = ''
df_sabin_covid['Ct_FluB'] = ''
df_sabin_covid['Ct_VSR'] = ''
df_sabin_covid['Ct_RDRP'] = ''
df_sabin_covid['Ct_geneE'] = ''
df_sabin_covid['Ct_geneN'] = ''
df_sabin_covid['Ct_geneS'] = ''
df_sabin_covid['Ct_ORF1ab'] = ''
df_sabin_covid['geneS_detection'] = ''
```

**Adding pathogens information**

```python
pathogens = [
    'SC2',
    'FLUA',
    'FLUB', 
    'VSR', 
    'META', 
    'RINO',
    'PARA', 
    'ADENO', 
    'BOCA', 
    'COVS', 
    'ENTERO', 
    'BAC'
]

for p in pathogens:
    if p != 'SC2':
        df_sabin_covid[p + '_test_result'] = 'NT' # 'Not tested'
```

```python
df_sabin_covid.insert(0, 'lab_id', 'SABIN')
```

```python
df_sabin_covid = df_sabin_covid.rename(columns=rename_cols_dict['SABIN'])
```

```python
rename_cols_dict = (
    df_fix_values
    .query('lab_id == "SABIN" or lab_id == "any"')
    .assign( mapping_data=lambda df: df.apply(lambda row: {row['old_data']: row['new_data']}, axis=1) )
    .groupby(['lab_id', 'column_name'])
    .agg({'mapping_data': lambda x: reduce(lambda a,b: {**a,**b}, x)})
    .reset_index()
    .assign( mapping=lambda df: df.apply(lambda row: {row['column_name']: row['mapping_data']}, axis=1) )
    .drop(columns=['mapping_data'])
    [['mapping']]
    .agg(lambda x: reduce(lambda a,b: {**a,**b}, x))
    .to_dict()
    ['mapping']
)

rename_cols_dict
```
