```python
# import combinations
from itertools import combinations, product

resultados = df_sabin['Resultado'].unique()
pathogens = [
    'FLUA','FLUB','VSR','META','RINO',
    'PARA','ADENO','BOCA','COVS','ENTERO','BAC'
]

pathogens_results = list(product (pathogens, resultados))
pathogens_results

df_pathogens_results = pd.DataFrame(pathogens_results, columns=['column_name', 'old_data'])
df_pathogens_results['column_name'] = df_pathogens_results['column_name'].apply(lambda x:  x + '_test_result')

df_pathogens_results['lab_id'] = 'SABIN'
df_pathogens_results['new_data'] = df_pathogens_results['old_data'].apply(
    lambda x: 'neg' if 'NÃO' in x else 'pos' 
) 

df_pathogens_results['new_data_2'] = df_pathogens_results['new_data'].apply(
    lambda x: 'Negative' if  x == 'neg' else 'Positive'
) 

df_pathogens_results
```
